```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv ('/Users/bartlomiejszafran/Desktop/archive/winemag-data_first150k.csv')
```

```python
data.head(15)
```

```python
sns.boxplot(data = data, y='price')
plt.show()
```

```python
max_data = data['price'].max()
min_data = data['price'].min()
mean_data = data['price'].mean()
median_data = data['price'].median()
std_data = data['price'].std()
print(max_data)
print(min_data)
print(mean_data)
print(median_data)
print(std_data)
```

```python
sns.histplot(data=data, x='price', kde='True')
plt.show()
```

```python
data.corr()
```

```python
sns.heatmap(data.corr(), cmap='coolwarm', annot=True)
plt.show()
```

```python
pivot_data = data.pivot_table(values='price', index='points', columns='country', fill_value=0)
print(pivot_data)
```

```python
filtered_data_over_80 = pivot_data[(data['points'] >= 80) & (data['points'] <= 100)]
print(filtered_data_over_80)
```

```python
#Graph is presenting the price valuse for vines with points over 80, also it is presenting its price. (If you have better suggestion how to do it, let me know)
sns.heatmap(filtered_data_over_80, vmin=10, vmax=2300)
plt.show()
```
